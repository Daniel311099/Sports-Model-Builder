```python
from bbc_sport_scraper import fetch 
import json
import time
import os
# load env variables
# from dotenv import load_dotenv
# load_dotenv()
DRIVE_PATH = os.getenv("DRIVE_PATH")
print(DRIVE_PATH)
```

```python
data = []
i = 2
attempt = 1
while True:
    time.sleep(10)
    print(f"fetching page {i}, attempt {attempt}")
    res = fetch(i)
    as_json = json.loads(res)
    commnts = as_json['payload'][0]['body']['results'] 
    print(f"got {len(commnts)} comments")
    if len(commnts) == 0:
        attempt += 1
    else:
        i += 1
    if attempt > 2:
        break
    data += commnts
    
```
